```python
import evdm
import numpy as np
import matplotlib.pyplot as plt
```

Create Instance of class "Body". 

In tis example rho function is 1-r,

the number op points (e.g. $r_i$) is 101,

velocity (is escape velocity on the edge of the body),

and dtype is type of body

```python
B = evdm.Body(lambda x: 1-x,101,velocity = 1e-3,Temp = 0,dtype = 'float')
```

We can plot some properties of body:
$\phi(r)$,$\rho(r)$,$M(r)$ $Q(r)$, where $$Q(r) = \cfrac{M(r)}{r^3}$$


```python
plt.plot(*B.phi)
plt.plot(*B.rho)
plt.plot(*B.Q)
plt.plot(*B.M)
plt.legend(["Phi(r)","Rho(r)","Q(r)","M(r)"])
```
